```python
def parse_json(json_data, row_dict, path=""):
    if isinstance(json_data, dict):
        for key, value in json_data.items():
            new_path = path + "." + key if path else key
            if isinstance(value, dict):
                parse_json(value, row_dict, new_path)
            else:
                row_dict[new_path] = value
    elif isinstance(json_data, list):
        for index, item in enumerate(json_data):
            new_path = path + f"[{index}]" if path else f"[{index}]"
            if isinstance(item, dict):
                parse_json(item, row_dict, new_path)
            else:
                row_dict[new_path] = item
    else:
        row_dict[path] = json_data
```
